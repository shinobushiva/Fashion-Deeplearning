```python
%matplotlib inline

from urllib.parse import urlparse
import matplotlib.pyplot as plt

import random
import shutil
import requests
import pickle
import os
from os.path import splitext
import re

from fashionutils import to_file_path
from PIL import Image
import numpy as np
import pandas as pd
```

```python
data_file = 'items.pickle'
img_folder = 'images'
out_folder = 'images-299'

items = {}
if os.path.exists(data_file):
    with open(data_file, mode='rb') as f:
        items = pickle.load(f)
```

```python

for k in items:
    item = items[k]
    # print(item)
    path = to_file_path(item['item_img_src'], item['item_href'], img_folder+"/items/")
    # print(path)
    item['item_category_simple'] = item['item_category'].split('(')[0]
    
```

```python
df = pd.DataFrame(list(items.values()))
df['item_category_simple_c'] = df['item_category_simple'].astype('category').cat.codes

# print(df.item_category_simple.value_counts())
# print(df.item_category_simple.value_counts().count())

pd.DataFrame(df.item_category_simple.value_counts())
```

```python
for i, v in df.iterrows():
    print(":"+v[3])
    if len(v[3]) <=0:
        continue;
    category = re.search('/category/(.*)/', v[3]).group(1)
    
    src_path = to_file_path(v[6], v[5], out_folder + '/items/').lower()
    dest_path = to_file_path(v[6], v[5], 'images-299-categorized/' + category + '/').lower()
    
    if not os.path.exists(src_path) or not os.path.exists(dest_path):
        continue;
    
#     print((os.path.abspath(src_path), os.path.abspath(dest_path)))
    shutil.copy(src_path, dest_path)
    
```

```python
IN_DIR = 'images-299-categorized'
TRAIN_DIR = 'images-finetune/train_images'
TEST_DIR = 'images-finetune/test_images'

if os.path.exists(TRAIN_DIR):
    shutil.rmtree(TRAIN_DIR)
os.makedirs(TRAIN_DIR)

if os.path.exists(TEST_DIR):
    shutil.rmtree(TEST_DIR)    
os.mkdir(TEST_DIR)

flower_dics = {}

list = os.listdir(IN_DIR)
for category in list:

    if len(os.listdir(os.path.join(IN_DIR, category))) >= 200:
        print(len(os.listdir(os.path.join(IN_DIR, category))))
        os.makedirs(os.path.join(TRAIN_DIR, category), exist_ok=True)
        os.makedirs(os.path.join(TEST_DIR, category), exist_ok=True)
```

```python
for f in sorted(os.listdir(IN_DIR)):
    files = os.listdir(os.path.join(IN_DIR, f))
    if len(files) >=200:
        for ff in files[10:200]:
            shutil.copy(os.path.join(IN_DIR, f, ff), os.path.join(TRAIN_DIR, f))
        for ff in files[0:10]:
            shutil.copy(os.path.join(IN_DIR, f, ff), os.path.join(TEST_DIR, f))
```
